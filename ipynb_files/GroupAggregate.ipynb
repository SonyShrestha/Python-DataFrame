```python
############### Read data from csv
import pandas as pd
df=pd.read_csv("D:\\data\\survey_results_public.csv")
df
```

```python
df.columns
```

```python
# Generate statistics on individual column
print(df["ResponseId"].min())
print(df["ResponseId"].max())
print(df["ResponseId"].mean())
print(df["ResponseId"].median())
print(df["ResponseId"].count())
print(df["ResponseId"].std())
```

```python
# Generate statistics on individual column
print(df["ResponseId"].describe())
```

```python
# Get statistics of all numeric fields
# Here, count is number of non NaN
df.describe()
```

```python
# Median of all numeric fields in a dataframe
df.median()
```

```python
# Count in each distinct category
df["Country"].value_counts()
```

```python
# Percentage in each distinct category
df["Country"].value_counts(normalize=True)
```

```python
# Get only country India
country_group=df.groupby(["Country"])
country_group.get_group("India")
```

```python
# Value count based on Sexuality for each country
country_group=df.groupby(["Country"])
country_group["Sexuality"].value_counts()
```

```python
# Value count based on Sexuality for specific country
country_group=df.groupby(["Country"])
country_group["Sexuality"].value_counts().loc["Afghanistan"]
```

```python
# Median for each group
country_group=df.groupby(["Country"])
country_group["CompTotal"].median()
```

```python
# Median and mean for each group
country_group=df.groupby(["Country"])
country_group["CompTotal"].agg(["median","mean"])
```

```python
# Median and mean for specific group
country_group=df.groupby(["Country"])
country_group["CompTotal"].agg(["median","mean"]).loc["Afghanistan"]
```

```python
# Get only country India and find number of values for each distinct category of Sexuality
fil=df["Country"]=="India"
df.loc[fil]["Sexuality"].value_counts()
```

```python
# Get only country India
fil=df["Country"]=="India"
df.loc[fil]
```

```python
# Aggregation based on group by 
df.groupby(["Country"]).agg(
    ResponseIdCount=("ResponseId","count"),
    ResponseIdSum=("ResponseId","sum"),
    AgeFirst=("Age","first"),
    MainBranchLast=("MainBranch","last")
        
).reset_index()
```
