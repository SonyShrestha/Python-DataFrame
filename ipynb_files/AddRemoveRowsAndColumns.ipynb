```python
# Create dataframe from dictionary
import pandas as pd
dict={
    "f_name":["Sony","Pragya","Prasanna","Ram","Hari","Sita"],
    "l_name":["Shrestha","Shrestha","Shrestha","Prasad","Kumar","Giri"],
    "age":[24,27,19,24,27,19]
}
df=pd.DataFrame(dict)
df
```

```python
# Add column
df["full_name"]=df["f_name"]+" "+df["l_name"]
df
```

```python
# Remove column
df.drop(columns=["f_name","l_name"],inplace=True)
df
```

```python
# Split
df["full_name"].str.split(" ")
```

```python
df["full_name"].str.split(" ",expand=True)
```

```python
df[["first_name","last_name"]]=df["full_name"].str.split(" ",expand=True)
```

```python
df
```

```python
# Add single row of data
df.append({"first_name":"Akhilesh","last_name":"Kumar"},ignore_index=True)
```
