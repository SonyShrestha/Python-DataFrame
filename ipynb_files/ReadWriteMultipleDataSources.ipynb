```python
############### Read data from csv with index_col=ResponseId
import pandas as pd
df=pd.read_csv("D:\\data\\survey_results_public.csv",index_col="ResponseId")
df
```

```python
# Filter records of Country=India and write to csv
india_df=df[df["Country"]=="India"]
india_df.to_csv("D:\\india_df.csv")
```

```python
# Filter records of Country=India and write to tsv
india_df=df[df["Country"]=="India"]
india_df.to_csv("D:\\india_df.tsv",sep='\t')
```
