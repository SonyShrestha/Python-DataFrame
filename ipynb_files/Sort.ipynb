```python
# Create dataframe from dictionary
import pandas as pd
dict={
    "f_name":["Sony","Pragya","Prasanna","Ram","Hari","Sita"],
    "l_name":["Shrestha","Shrestha","Shrestha","Prasad","Kumar","Giri"],
    "age":[24,27,19,24,27,19]
}
df=pd.DataFrame(dict)
df
```

```python
# Sort rows by first name in ascending order
df.sort_values(by="f_name")
```

```python
# Sort rows by first name in descending order
df.sort_values(by="f_name",ascending=False)
```

```python
# Sort rows by last name then first name in descending order
df.sort_values(by=["l_name","f_name"],ascending=False)
```

```python
# Sort rows by last name then first name in ascending and descending order respectively
df.sort_values(by=["l_name","f_name"],ascending=[True,False])
```

```python
df
```

```python
# Make changes made by sort permanent
df.sort_values(by=["l_name","f_name"],ascending=[True,False],inplace=True)
df
```

```python
# Reset the sorted dataframe
df.sort_index(inplace=True)
df
```

```python
# Sort specific column
df["l_name"].sort_values()
```

```python
# Sort specific column
df[["f_name","l_name"]].sort_values(by="f_name")
```

```python
# Get 3 highest salary
df["age"].nlargest(3)
```

```python
# Get 3 highest salary records
df.nlargest(3,"age")
```

```python
# Get 3 smallest salary
df["age"].nsmallest(3)
```

```python
# Get 3 smallest salary records
df.nsmallest(3,"age")
```
